```python
import torch
from torch.utils.data import Dataset
import torch.nn.functional as F
import torch.nn as nn
import numpy as np
```

```python
x = []
y = []
for i in range(5):
    x.append([float(i),float(i+1)])
    y.append(0)
x = x*50
y = y*50
x1 = []
y1 = []
for i in range(1,6):
    x1.append([float(i),float(i-1)])
    y1.append(1)
x1 = x1*50
y1 = y1*50
x = x1+x
y = y1+y
print(len(x))

print(len(y))
```

```python
x = np.array(x,dtype=float)
y = np.array(y,dtype=float)
x = torch.from_numpy(x)
y = torch.from_numpy(y)
x.dtype
x
```

```python
N, D_in, H1, H2, D_out = 10, 2, 30, 15, 1
dtype = torch.float64
device = torch.device("cpu")
lr = .01
epochs = 1
```

```python
w1 = torch.randn(D_in, H1, device=device, dtype=dtype, requires_grad=True)
#b1 = torch.randn(D, H1, device=device, dtype=dtype, requires_grad=True)
w2 = torch.randn(H1, H2, device=device, dtype=dtype, requires_grad=True)
w3 = torch.randn(H2, D_out, device=device, dtype=dtype, requires_grad=True)
w1.dtype
#print(w1)
```

```python
for epoch in range(1000):
    # forward pass
    runningLoss = 0
    yhat = F.relu(F.relu(x.mm(w1)).mm(w2)).mm(w3)
    #print(yhat.shape)
    #print(y.shape)
    loss = (yhat - y).pow(2).sum()
    runningLoss += loss.item()
    loss.backward()
    #print(w1)
    with torch.no_grad():
        w1 -= lr *w1.grad
        w2 -= lr *w2.grad
        w3 -= lr *w3.grad
        w1.grad.zero_()
        w2.grad.zero_()
        
print(w1)
```
